```python
import pandas as pd

#Sitemap was taken from https://www.activateuts.com.au/sitemap_index.xml and converted to CSV
df = pd.read_csv('/content/drive/MyDrive/datasets/sitemap.csv')
df.head()
```

```python
df['formatted'] = df['loc'].apply(lambda x: f'"{x.split("/")[-2]}":"{"/" + "/".join(x.split("/")[3:])[:-1]}",')
df[['formatted']].head()
```

```python
from google.colab import files
df[['formatted']].to_csv('formatted_data.csv', index=False)
files.download('formatted_data.csv')
```
